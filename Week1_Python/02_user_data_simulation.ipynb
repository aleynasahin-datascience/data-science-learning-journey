# User Data Simulation Project

## Purpose
Simulating user dataset and performing preprocessing operations for future machine learning tasks.


```python
# User List
users = ["Ali", "Ayşe", "Mehmet", "Zeynep", "Can", "Elif"]
print(users)
```

```python
# Ages
ages = {
    "Ali": 23,
    "Ayşe": 31,
    "Mehmet": 28,
    "Zeynep": 35,
    "Can": 19,
    "Elif": 42
}
print(ages)
```

```python
# Purchase Amount
purchase_amount = {
    "Ali": 120,
    "Ayşe": 540,
    "Mehmet": 300,
    "Zeynep": 700,
    "Can": 90,
    "Elif": 850
}
print(purchase_amount)
```

```python
young_users = [user for user in users if ages[user] < 30]
print(young_users)
```

```python
high_value_users = [user for user in users if purchase_amount[user] > 400]
print(high_value_users)
```

```python
user_segment = {
    user: "Young" if ages[user] < 30 else "Adult"
    for user in users
}
print(user_segment)
```

```python
average_purchase = sum(purchase_amount.values()) / len(purchase_amount)
print(average_purchase)
```

```python
above_average_users = [
    user for user in users
    if purchase_amount[user] > average_purchase
]
print(above_average_users)
```
