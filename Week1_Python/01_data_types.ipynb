# Python Veri Tipleri

## Amaç
Bu çalışma Python temel veri tiplerini öğrenmek ve veri bilimi kullanımını anlamak için hazırlanmıştır.

## Öğrenilecek Konular
- Integer
- Float
- String
- Boolean
- type() fonksiyonu


```python
x = 10
print(x)
print(type(x))
```

```python
y = 10.5
print(y)
print(type(y))
```

```python
name = "Aleyna"
print(name)
print(type(name))
```

```python
status = True
print(status)
print(type(status))
```

## Mini Proje — Kullanıcı Bilgileri Veri Tipi Analizi


```python
age = 24
weight = 1.58
city = "Ankara"
is_working = False

print(type(age))
print(type(weight))
print(type(city))
print(type(is_working))
```
